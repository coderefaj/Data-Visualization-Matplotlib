## Data Visualization :

Data Visualization allows us to quickly intrepret the data and adjust different variables to see their effects . Since human brains are easy to analyze the visual than text . 

Also it helps Non-Tech people to understand about our analysis . 

Data visualization is the presentation of data in a pictorial or graphical format . 



```python
from matplotlib import pyplot as plt
plt.plot([1,2,3],[4,5,1]) # X , Y 
plt.show()
```

```python
x = [5,8,10]
y = [12,16,6]

plt.plot(x,y)
plt.title('Info')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')

x = [5,8,10]
y = [12,16,6]

x2 = [6,9,11]
y2 = [6,15,7]

plt.plot(x,y,'g',label="line one",linewidth=5)
plt.plot(x2,y2,'b',label="line two",linewidth=5)

plt.title("Epic Info")
plt.ylabel("Y axis")
plt.xlabel("X axis")

plt.legend()

plt.grid(True , color="k")
# https://matplotlib.org/api/_as_gen/matplotlib.pyplot.colors.html

plt.show()
```

Bar graph used to compare things between different groups . 

```python
x = [1,2,3,4,5]
y = [6,5,7,2,9]

x2 = [2,4,6,8,10]
y2 = [1,3,5,7,9]

plt.bar(x,y,label="Example one",color='r')
plt.bar(x2,y2,label="Example two",color='g')

plt.legend()
plt.xlabel("Bar number")
plt.ylabel("Bar height")
plt.title("Info")

plt.show()
```
